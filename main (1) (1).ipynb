```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB   
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN 

def preprocess(df):
    lemmatizer = WordNetLemmatizer()

    stop_words = set(stopwords.words('english'))
    stop_words = stop_words | set(["na", "oh", "ooh", "ta", "uh", "na", "ey", "yeah", "ya", "aw"])
    preprocessed_texts = []

    for text in df['text']:
        text = re.sub('[^a-zA-Z]', ' ', text)
        text = re.sub(r'\s+', ' ', text)

        tokens = word_tokenize(text.lower())

        tokens = [
            token for token in tokens
            if token not in stop_words and len(token) > 1
        ]

        tagged = pos_tag(tokens)

        lemmas = [
            lemmatizer.lemmatize(token, get_wordnet_pos(tag))
            for token, tag in tagged
        ]

        preprocessed_texts.append(" ".join(lemmas))

    return pd.DataFrame({"text": preprocessed_texts})
```

```python
def filter_by_tfidf(df, tfidf_df):
    filtered_texts = []

    for i in range(df.shape[0]):
        original_words = df['text'].iloc[i].split()
        tfidf_row = tfidf_df.iloc[i]
        allowed_words = set(tfidf_row[tfidf_row > 0].index)

        filtered = [w for w in original_words if w in allowed_words]
        filtered_texts.append(" ".join(filtered))

    df['tfidf_filtered_text'] = filtered_texts
    return df
```

```python


def get_bigr_trigr_from_filtered(df, top_n=40):

    big_list = []
    tri_list = []

    for song_idx, text in enumerate(df["tfidf_filtered_text"]):
        words = text.split()
        big_list += [(" ".join(bg), song_idx) for bg in bigrams(words)]
        tri_list += [(" ".join(tri), song_idx) for tri in trigrams(words)]

    big_freq = Counter(big_list)
    tri_freq = Counter(tri_list)

    big_df = pd.DataFrame(
        [[ngram, song_idx, n] for (ngram, song_idx), n in big_freq.most_common(top_n)],
        columns=["bigr", "song", "n"]
    )

    tri_df = pd.DataFrame(
        [[ngram, song_idx, n] for (ngram, song_idx), n in tri_freq.most_common(top_n)],
        columns=["bigr", "song", "n"]
    )

    return big_df, tri_df


def draw_net(big_df, tri_df,
             big_color='#e37171',
             tri_color='#e37171',
             node_palette=None):
    G = nx.Graph()

    if node_palette is None:
        node_palette = [
            '#1f77b4', 
            '#ff7f0e',
            '#2ca02c',
            '#d62728',
            '#9467bd',  
        ]

    for _, row in big_df.iterrows():
        w = row['bigr'].split()
        song_idx = int(row['song'])
        G.add_edge(
            w[0], w[1],
            weight=row['n'],
            type='bigram',
            song=song_idx
        )

    for _, row in tri_df.iterrows():
        w = row['bigr'].split() 
        song_idx = int(row['song'])
        # два ребра: word1-word2, word2-word3
        G.add_edge(
            w[0], w[1],
            weight=row['n'],
            type='trigram',
            song=song_idx
        )
        G.add_edge(
            w[1], w[2],
            weight=row['n'],
            type='trigram',
            song=song_idx
        )

    from collections import defaultdict

    node_song_counts = defaultdict(lambda: Counter())

    for u, v, data in G.edges(data=True):
        song_idx = data.get('song', 0)
        node_song_counts[u][song_idx] += 1
        node_song_counts[v][song_idx] += 1

    node_main_song = {}
    for node, c in node_song_counts.items():
        main_song = max(c.items(), key=lambda x: x[1])[0]
        node_main_song[node] = main_song

    layout = nx.spring_layout(G, k=0.35, seed=42)

    edge_traces = []

    edge_x = []
    edge_y = []
    for u, v, data in G.edges(data=True):
        if data['type'] == 'bigram':
            x0, y0 = layout[u]
            x1, y1 = layout[v]
            edge_x += [x0, x1, None]
            edge_y += [y0, y1, None]

    edge_traces.append(
        go.Scatter(
            x=edge_x, y=edge_y,
            mode='lines',
            line=dict(width=1, color=big_color),
            name='bigrams',
            hoverinfo='none'
        )
    )

    edge_x = []
    edge_y = []
    for u, v, data in G.edges(data=True):
        if data['type'] == 'trigram':
            x0, y0 = layout[u]
            x1, y1 = layout[v]
            edge_x += [x0, x1, None]
            edge_y += [y0, y1, None]

    edge_traces.append(
        go.Scatter(
            x=edge_x, y=edge_y,
            mode='lines',
            line=dict(width=1, color=tri_color),
            name='trigrams',
            hoverinfo='none'
        )
    )

    node_x = []
    node_y = []
    node_labels = []
    node_colors = []

    for node in G.nodes():
        x, y = layout[node]
        node_x.append(x)
        node_y.append(y)
        node_labels.append(node)

        song_idx = node_main_song.get(node, 0)
        color = node_palette[song_idx % len(node_palette)]
        node_colors.append(color)

    nodes = go.Scatter(
        x=node_x, y=node_y,
        mode='markers+text',
        text=[f"{lbl}" for lbl in node_labels],
        textposition='top center',
        marker=dict(size=10, color=node_colors),
        name='words',
        hoverinfo='text'
    )

    fig = go.Figure(
        data=edge_traces + [nodes],
        layout=go.Layout(
            width=1300,
            height=1000,
            showlegend=True,
            hovermode="closest",
            margin=dict(l=0, r=0, t=40, b=0),
            title="Bigram + Trigram Network"
        )
    )

    fig.show()
```

```python
class IMDBDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=128):
        self.texts = list(texts)
        self.labels = list(labels)
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = int(self.labels[idx])

        encoding = self.tokenizer(
            text,
            max_length=self.max_len,
            padding='max_length',  
            truncation=True,      
            return_tensors='pt',
            return_attention_mask=True
        )

        item = {
            'input_ids': encoding['input_ids'].squeeze(0),         
            'attention_mask': encoding['attention_mask'].squeeze(0),
            'labels': torch.tensor(label, dtype=torch.long)
        }

        if 'token_type_ids' in encoding:
            item['token_type_ids'] = encoding['token_type_ids'].squeeze(0)

        return item
```

```python

MAX_LEN = 256
BATCH_SIZE = 16

train_dataset = IMDBDataset(X_train, y_train, tokenizer, max_len=MAX_LEN)
test_dataset  = IMDBDataset(X_test,  y_test,  tokenizer, max_len=MAX_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False)
```

```python
import datetime
from tqdm import tqdm  # красивый прогресс-бар (если не стоит: pip install tqdm)
from transformers import AdamW, get_cosine_schedule_with_warmup

epochs = 4
optimizer = AdamW(model.parameters(), lr=2e-5)

total_steps = len(train_loader) * epochs
num_warmup_steps = int(0.1 * total_steps)  # 10% шагов на warmup

scheduler = get_cosine_schedule_with_warmup(
    optimizer,
    num_warmup_steps=num_warmup_steps,
    num_training_steps=total_steps
)

start_time = datetime.datetime.now()
print("Training started:", start_time)

global_step = 0

for epoch in range(epochs):
    model.train()
    total_loss = 0
    
    print(f"\n===== Epoch {epoch+1}/{epochs} =====")
    print(f"Current LR at epoch start: {optimizer.param_groups[0]['lr']:.6f}")
    
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}")

    for step, batch in enumerate(progress_bar, start=1):
        optimizer.zero_grad()

        batch = {k: v.to(device) for k, v in batch.items()}

        outputs = model(
            input_ids=batch['input_ids'],
            attention_mask=batch['attention_mask'],
            labels=batch['labels']
        )

        loss = outputs.loss
        loss.backward()
        optimizer.step()
        scheduler.step()  # обновляем LR по cosine + warmup

        total_loss += loss.item()
        global_step += 1

        progress_bar.set_postfix({
            "batch_loss": loss.item(),
            "lr": optimizer.param_groups[0]['lr']
        })

    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch+1} finished. Average loss = {avg_loss:.4f}")
    print(f"LR at epoch end: {optimizer.param_groups[0]['lr']:.6f}")

end_time = datetime.datetime.now()
print("\nTraining finished:", end_time)
print("Duration:", end_time - start_time)
```

```python
model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for batch in test_loader:
        labels = batch['labels'].to(device)
        batch = {k: v.to(device) for k, v in batch.items()}

        outputs = model(
            input_ids=batch['input_ids'],
            attention_mask=batch['attention_mask']
        )

        logits = outputs.logits
        preds = torch.argmax(logits, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

acc = accuracy_score(all_labels, all_preds)
precision, recall, f1, _ = precision_recall_fscore_support(
    all_labels, all_preds, average='binary'
)

print(f'Accuracy:  {acc:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall:    {recall:.4f}')
print(f'F1-score:  {f1:.4f}')
```
